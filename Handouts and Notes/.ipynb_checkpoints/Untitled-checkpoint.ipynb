```python
from scipy.stats import norm
df=pd.DataFrame(norm(0,1).rvs(size=(1000,100)))
df
```

```python
def test_loc(df):
    for i in df.index:
        val = df.loc[i,0]
%timeit test_loc(df)
```

```python
def test_iterrow(df):
    for (i, row) in df.iterrows():
        val = row.loc[0]
%timeit test_iterrow(df)
```

```python
def test_loc(df):
    for i in df.index:
        val = df.at[i,0]
%timeit test_loc(df)
```
